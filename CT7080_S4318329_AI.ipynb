```python
def convert_dummy_variables(df):
    """
    Convert categorical variables into dummy variables.
    
    Args:
    - df: DataFrame containing categorical variables to be converted
    
    Returns:
    - df: DataFrame with dummy variables and original categorical variables dropped
    """
    # Convert protocol_type into dummy variables
    df["is_icmp"] = df["protocol_type"].apply(lambda x: 1 if x == "icmp" else 0)
    df["is_tcp"] = df["protocol_type"].apply(lambda x: 1 if x == "tcp" else 0)
    df["is_udp"] = df["protocol_type"].apply(lambda x: 1 if x == "udp" else 0)
    
    # Convert service into dummy variables
    df["is_ecr_i"] = df["service"].apply(lambda x: 1 if x == "ecr_i" else 0)
    df["is_private"] = df["service"].apply(lambda x: 1 if x == "private" else 0)
    df["is_http"] = df["service"].apply(lambda x: 1 if x == "http" else 0)
    df["is_smtp"] = df["service"].apply(lambda x: 1 if x == "smtp" else 0)
    df["is_domain_u"] = df["service"].apply(lambda x: 1 if x == "domain_u" else 0)
    df["is_ftp_data"] = df["service"].apply(lambda x: 1 if x == "ftp_data" else 0)
    
    # Convert flag into dummy variables
    df["is_SF"] = df["flag"].apply(lambda x: 1 if x == "SF" else 0)
    df["is_S0"] = df["flag"].apply(lambda x: 1 if x == "S0" else 0)
    df["is_REJ"] = df["flag"].apply(lambda x: 1 if x == "REJ" else 0)
    
    # Drop original categorical variables
    df = df.drop(["protocol_type", "service", "flag"], axis=1)
    
    return df
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
import numpy as np
import time

class ClassifierEvaluation:
    def __init__(self, X_train, y_train, X_test, y_test):
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test

    def train_random_forest(self):
        self.rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
        self.rf_classifier.fit(self.X_train, self.y_train)

    def train_naive_bayes(self):
        self.nb_classifier = GaussianNB()
        self.nb_classifier.fit(self.X_train, self.y_train)

    def predict_proba_random_forest(self):
        self.y_score_rf = self.rf_classifier.predict_proba(self.X_test)

    def predict_proba_naive_bayes(self):
        self.y_score_nb = self.nb_classifier.predict_proba(self.X_test)

    def plot_precision_recall_curves(self):
        fig, axes = plt.subplots(2, 4, figsize=(15, 6))

        classifiers = [("Random Forest", self.y_score_rf), ("Naive Bayes", self.y_score_nb)]
        colors = ['b', 'g', 'r', 'c']

        for clf_name, y_score in classifiers:
            precision = dict()
            recall = dict()
            average_precision = dict()

            for i, color in zip(range(4), colors):
                y_test_binary = label_binarize(self.y_test, classes=np.arange(4))  # Exclude Class 4
                precision[i], recall[i], _ = precision_recall_curve(y_test_binary[:, i], y_score[:, i])
                average_precision[i] = average_precision_score(y_test_binary[:, i], y_score[:, i])

                axes[classifiers.index((clf_name, y_score))][i].plot(recall[i], precision[i], color=color, lw=2, label=f'Class {i} (AUC-PR = {average_precision[i]:.2f})')
                axes[classifiers.index((clf_name, y_score))][i].set_xlabel('Recall')
                axes[classifiers.index((clf_name, y_score))][i].set_ylabel('Precision')
                axes[classifiers.index((clf_name, y_score))][i].set_title(f'{clf_name} - Class {i}')
                axes[classifiers.index((clf_name, y_score))][i].legend()

        plt.tight_layout()
        plt.show()

    def evaluate_elapsed_time(self):
        start_time = time.time()

        self.train_random_forest()
        self.train_naive_bayes()
        self.predict_proba_random_forest()
        self.predict_proba_naive_bayes()
        self.plot_precision_recall_curves()

        end_time = time.time()
        elapsed_time = end_time - start_time
        print(f"Elapsed time: {elapsed_time:.2f} seconds")

# Usage
if __name__ == "__main__":
    # Assuming X_train, y_train, X_test, and y_test are defined
    classifier_eval = ClassifierEvaluation(X_train, y_train, X_test, y_test)
    classifier_eval.evaluate_elapsed_time()
```
